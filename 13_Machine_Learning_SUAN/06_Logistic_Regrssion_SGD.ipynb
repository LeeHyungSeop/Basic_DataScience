# Logistic Regression

* 로지스틱 회귀는 회귀라는 이름이 붙어 있지만, 사실은 분류에 사용되는 선형 모델이다
* 로지스틱 회귀는 매우 강력한 분류 알고리즘이다.
* 일반적으로 이진 분류 뿐만 아니라 희소 영역의 분류, 예를 들어 텍스트 분류와 같은 영역에서 뛰어난 예측 성능을 보인다.
* $\sigma$ : 시그모이드 함수


* 로지스틱 회귀 모델은 선형 회귀 모델에 시그모이드 함수 적용
* 로지스틱 회귀의 학습 목표는 다음과 같은 목적 함수를 최소화하는 파라미터 $\omega$를 찾는 것.


# Logistic Regression 예제

이진 분류

## 이진 데이터 생성

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(['seaborn-whitegrid'])
```

```python
from sklearn.datasets import make_classification # 데이터를 직접 만들기 위해서
from sklearn.model_selection import train_test_split, cross_val_score 
from sklearn.linear_model import LogisticRegression
```

```python
samples = 1000 # sample 개수 1,000개
X, y = make_classification(n_samples=samples, n_features=2,  # binary data
                           n_informative=2, n_redundant=0,   # noise 없게
                           n_clusters_per_class=1) # class당 cluster 1개
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

ax.grid()
ax.set_xlabel('X')
ax.set_ylabel('y')

for i in range(samples) :
    if y[i] == 0 : # 0이면
        ax.scatter(X[i, 0], X[i, 1], edgecolors='k', alpha=0.5, marker='^', color='r')
    else :         # 1이면
        ax.scatter(X[i, 0], X[i, 1], edgecolors='k', alpha=0.5, marker='v', color='b')

plt.show()
```

## 분류를 위한 Logistic Regression 모델 생성

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
print("학습 데이터 점수 : {}" .format(model.score(X_train, y_train)))
print("평가 데이터 점수 : {}" .format(model.score(X_test, y_test)))

# Logistic Regression은 매우 잘 분류되는 모델이다.
```

## cross_val_score

```python
scores = cross_val_score(model, X, y, scoring='accuracy', cv=10)
print("CV 평균 점수 : {}" .format(scores.mean()))
```

```python
model.intercept_, model.coef_
```

## Logistic Regression 결과 시각화

```python
x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
y_min, y_max = X[:,1].min() - 0.5, X[:,1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(1, figsize=(10, 6))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Pastel1)
plt.scatter(X[:,0], X[:,1], c=np.abs(y-1), edgecolors='k', alpha=0.5, cmap=plt.cm.coolwarm)
plt.xlabel('X')
plt.ylabel('y')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks()
plt.yticks()
plt.show()

# 두 가지 클래스를 구분하는 경계선 -> Logistic Regression의 결과 -> 얼만큼 잘 분류했는지
```

# 붓꽃 데이터에 대한 Logistic Regression

## 붓꽃 데이터셋 로드 및 탐색

```python
from sklearn.datasets import load_iris

iris = load_iris()
print(iris.keys())
print(iris.DESCR)

''' 
:Attribute Information:
        - sepal length in cm
        - sepal width in cm
        - petal length in cm
        - petal width in cm
        - class:
                - Iris-Setosa
                - Iris-Versicolour
                - Iris-Virginica
'''
# feature : 4개
# sepal, petal 정보로 3개의 class 중 어떤 것인지 맞추는 분류 문제.
# -> class가 3개 -> binary는 아님 -> 분류할 수 있는 정답이 3개
# sepal : 꽃받침
# petal : 꽃잎
```

```python
import pandas as pd 

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
species = pd.Series(iris.target, dtype='category')
species = species.cat.rename_categories(iris.target_names)
iris_df['species'] = species

iris_df.describe()
```

```python
iris_df.boxplot();

# sepal 값 범위는 크지 않고, outlier(이상치) 발견됨
# petal 값 범위는 큰 것 같음 -> 데이터가 넓게 퍼져있는 형태.
```

```python
iris_df.plot();

# 0~50 , 50~100, 101~ 까지의 그룹으로 되어 있는 것을 확인할 수 있다.
```

```python
import seaborn as sns

sns.pairplot(iris_df, hue='species')
# feature가 4개 -> 4 * 4

# 파란색 (setosa)에 대해서는 분류가 쉬워보임.
# 주황색(versicolor), 초록색(virginica)은 겹치는 것이 보여 분류가 상대적으로 어려워 보임.
```

## 붓꽃 데이터에 대한 Logistic Regression

```python
from sklearn.model_selection import train_test_split

# feature 2, 3 == petal length (cm)	petal width (cm) 에 대해서만 사용
X_train, X_test, y_train, y_test = train_test_split(iris.data[:, [2, 3]], iris.target,
                                                    test_size=0.2, random_state=1, stratify=iris.target) # stratify : 계층적 구조를 가진 data를 split할 때 좋다 
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', multi_class='auto', C=100.0, random_state=1)
model.fit(X_train, y_train)
```

```python
print("학습 데이터 점수 : {}" .format(model.score(X_train, y_train)))
print("평가 데이터 점수 : {}" .format(model.score(X_test, y_test)))
```

## 분류 시각화

```python
import numpy as np

X = np.vstack([X_train, X_test]) # X 훈련용, 학습용 데이터 묶기
y = np.hstack([y_train, y_test]) # y 훈련용, 학습용 데이터 묶기
```

```python
from matplotlib.colors import ListedColormap

x1_min, x1_max = X[:, 0].min()-1, X[:, 0].max()+1
x2_min, x2_max = X[:, 0].min()-1, X[:, 1].max()+1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02),
                       np.arange(x2_min, x2_max, 0.02))

Z = model.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
Z = Z.reshape(xx1.shape)

species = ('Setosa', 'Versicolour', 'Virginica')
markers = ('^', 'v', 's')
colors = ('blue', 'orange', 'green')
cmap = ListedColormap(colors[:len(np.unique(y))])
plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
plt.xlim(xx1.min(), xx1.max())
plt.ylim(xx2.min(), xx2.max())

for idx, cl in enumerate(np.unique(y)) :
    plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                alpha=0.8, c=colors[idx],
                marker=markers[idx], label=species[cl],
                edgecolors='k')

# 아까 train, test 데이터를 합쳤는데 다시 나누는 작업. 임의의 영역을 뽑아서 test 부분을 가져온다.
X_comb_test, y_comb_test = X[range(105, 150), :], y[range(105, 150)] 
plt.scatter(X_comb_test[:, 0], X_comb_test[:, 1],
            c='red', edgecolors='k', alpha=0.4,
            linewidths=1, marker='o', s=100, label='test')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(loc='upper left')
plt.tight_layout();
```

## GridSearchCV : 최적 파라미터 찾아 성능 높히기

```python
import multiprocessing
from sklearn.model_selection import GridSearchCV

param_grid = [
    {
        'penalty' : ['l1', 'l2'],
        'C' : [2.0, 2.2, 2.4, 2.6, 2.8]
    }
]

gs = GridSearchCV(estimator=LogisticRegression(), param_grid=param_grid,
                  scoring='accuracy', cv=10, n_jobs=multiprocessing.cpu_count())       
gs
```

```python
result = gs.fit(iris.data, iris.target)
```

```python
print(gs.best_estimator_)
print("최적 점수 : {}" .format(gs.best_score_))
print("최적 파라미터 : {}" .format(gs.best_params_))

pd.DataFrame(result.cv_results_)
```

# 확률적 경사 하강법 (Stochastic Gradient Descent, SGD) 

* 모델을 학습시키기 위한 간단한 방법
* 학습 파라미터에 대한 손실 함수(비용 함수)의 기울기를 구해 기울기가 최소화 되는 방향으로 학습

* scikit-learn에서는 선형 SGD 회귀와 SGD 분류를 지원

## SGD를 사용한 선형 회귀 분석

## boston dataset load

```python
from sklearn.linear_model import SGDRegressor # SGD 회귀
from sklearn.datasets import _california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

import pandas as pd
import numpy as np

boston_df = pd.read_csv(
    filepath_or_buffer="http://lib.stat.cmu.edu/datasets/boston",
    delim_whitespace=True,
    skiprows=21,
    header=None,
)

columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']

#Flatten all the values into a single long list and remove the nulls
values_w_nulls = boston_df.values.flatten()
all_values = values_w_nulls[~np.isnan(values_w_nulls)]

#Reshape the values to have 14 columns and make a new boston_df out of them
boston_df = pd.DataFrame(
    data = all_values.reshape(-1, len(columns)),
    columns = columns,
)

boston_df
```

```python
X, y = boston_df.iloc[:, :-1], boston_df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

## loss function 지정해주기

```python
model = make_pipeline(
    StandardScaler(),
    SGDRegressor(loss='squared_error')  # 선형회귀할 때는 squared_error를 주로 사용
)
model.fit(X_train, y_train)
```

```python
print("학습 데이터 점수 : {}" .format(model.score(X_train, y_train)))
print("평가 데이터 점수 : {}" .format(model.score(X_test, y_test))) 
```

# 붓꽃 데이터에 대한 SGD 분류

```python
from sklearn.linear_model import SGDClassifier # SGD 분류
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# SGDClassifer()는 default로, loss='hing' -> support vector machine을 사용하게 되어있는데, 
# Support Vector Machine을 아직 안배웠고 Logistic Regressionloss을 배우고 있으니까 loss='log'로 지정
model = make_pipeline(StandardScaler(), SGDClassifier(loss='log')) 
model.fit(X_train, y_train) 
```

```python
print("학습 데이터 점수 : {}" .format(model.score(X_train, y_train)))
print("평가 데이터 점수 : {}" .format(model.score(X_test, y_test)))
```

# 유방암 데이터에 대한 SGD 분류

```python
from sklearn.linear_model import SGDClassifier # SGD 분류
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# SGDClassifer()는 default로, loss='hing' -> support vector machine을 사용하게 되어있는데, 
# Support Vector Machine을 아직 안배웠고 Logistic Regressionloss을 배우고 있으니까 loss='log'로 지정
model = make_pipeline(StandardScaler(), SGDClassifier(loss='log')) 
model.fit(X_train, y_train) 
```

```python
print("학습 데이터 점수 : {}" .format(model.score(X_train, y_train)))
print("평가 데이터 점수 : {}" .format(model.score(X_test, y_test)))
```
