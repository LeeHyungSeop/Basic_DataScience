```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(['seaborn-whitegrid'])
```

```python
from sklearn.datasets import make_classification #    
from sklearn.model_selection import train_test_split, cross_val_score 
from sklearn.linear_model import LogisticRegression
```

```python
samples = 1000 # sample  1,000
X, y = make_classification(n_samples=samples, n_features=2,  # binary data
                           n_informative=2, n_redundant=0,   # noise 
                           n_clusters_per_class=1) # class cluster 1
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

ax.grid()
ax.set_xlabel('X')
ax.set_ylabel('y')

for i in range(samples) :
    if y[i] == 0 : # 0
        ax.scatter(X[i, 0], X[i, 1], edgecolors='k', alpha=0.5, marker='^', color='r')
    else :         # 1
        ax.scatter(X[i, 0], X[i, 1], edgecolors='k', alpha=0.5, marker='v', color='b')

plt.show()
```

##   Logistic Regression  

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
print("   : {}" .format(model.score(X_train, y_train)))
print("   : {}" .format(model.score(X_test, y_test)))

# Logistic Regression    .
```

## cross_val_score

```python
scores = cross_val_score(model, X, y, scoring='accuracy', cv=10)
print("CV   : {}" .format(scores.mean()))
```

```python
model.intercept_, model.coef_
```

## Logistic Regression  

```python
x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
y_min, y_max = X[:,1].min() - 0.5, X[:,1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(1, figsize=(10, 6))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Pastel1)
plt.scatter(X[:,0], X[:,1], c=np.abs(y-1), edgecolors='k', alpha=0.5, cmap=plt.cm.coolwarm)
plt.xlabel('X')
plt.ylabel('y')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks()
plt.yticks()
plt.show()

#      -> Logistic Regression  ->   
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
print(iris.keys())
print(iris.DESCR)

''' 
:Attribute Information:
        - sepal length in cm
        - sepal width in cm
        - petal length in cm
        - petal width in cm
        - class:
                - Iris-Setosa
                - Iris-Versicolour
                - Iris-Virginica
'''
# feature : 4
# sepal, petal  3 class      .
# -> class 3 -> binary  ->     3
# sepal : 
# petal : 
```

```python
import pandas as pd 

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
species = pd.Series(iris.target, dtype='category')
species = species.cat.rename_categories(iris.target_names)
iris_df['species'] = species

iris_df.describe()
```

##    Logistic Regression

```python
from sklearn.model_selection import train_test_split

# feature 2, 3 == petal length (cm)	petal width (cm)   
X_train, X_test, y_train, y_test = train_test_split(iris.data[:, [2, 3]], iris.target,
                                                    test_size=0.2, random_state=1, stratify=iris.target) # stratify :    data split   
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', multi_class='auto', C=100.0, random_state=1)
model.fit(X_train, y_train)
```

```python
print("   : {}" .format(model.score(X_train, y_train)))
print("   : {}" .format(model.score(X_test, y_test)))
```

##  

```python
import numpy as np

X = np.vstack([X_train, X_test]) # X ,   
y = np.hstack([y_train, y_test]) # y ,   
```

```python
from matplotlib.colors import ListedColormap

x1_min, x1_max = X[:, 0].min()-1, X[:, 0].max()+1
x2_min, x2_max = X[:, 0].min()-1, X[:, 1].max()+1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02),
                       np.arange(x2_min, x2_max, 0.02))

Z = model.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
Z = Z.reshape(xx1.shape)

species = ('Setosa', 'Versicolour', 'Virginica')
markers = ('^', 'v', 's')
colors = ('blue', 'orange', 'green')
cmap = ListedColormap(colors[:len(np.unique(y))])
plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
plt.xlim(xx1.min(), xx1.max())
plt.ylim(xx2.min(), xx2.max())

for idx, cl in enumerate(np.unique(y)) :
    plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                alpha=0.8, c=colors[idx],
                marker=markers[idx], label=species[cl],
                edgecolors='k')

#  train, test     .    test  .
X_comb_test, y_comb_test = X[range(105, 150), :], y[range(105, 150)] 
plt.scatter(X_comb_test[:, 0], X_comb_test[:, 1],
            c='red', edgecolors='k', alpha=0.4,
            linewidths=1, marker='o', s=100, label='test')

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(loc='upper left')
plt.tight_layout();
```

```python
from sklearn.linear_model import SGDRegressor # SGD 
from sklearn.datasets import _california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

import pandas as pd
import numpy as np

boston_df = pd.read_csv(
    filepath_or_buffer="http://lib.stat.cmu.edu/datasets/boston",
    delim_whitespace=True,
    skiprows=21,
    header=None,
)

columns = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']

#Flatten all the values into a single long list and remove the nulls
values_w_nulls = boston_df.values.flatten()
all_values = values_w_nulls[~np.isnan(values_w_nulls)]

#Reshape the values to have 14 columns and make a new boston_df out of them
boston_df = pd.DataFrame(
    data = all_values.reshape(-1, len(columns)),
    columns = columns,
)

boston_df
```

```python
X, y = boston_df.iloc[:, :-1], boston_df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

## loss function 

```python
model = make_pipeline(
    StandardScaler(),
    SGDRegressor(loss='squared_error')  #   squared_error  
)
model.fit(X_train, y_train)
```

```python
print("   : {}" .format(model.score(X_train, y_train)))
print("   : {}" .format(model.score(X_test, y_test))) 
```

#    SGD 

```python
from sklearn.linear_model import SGDClassifier # SGD 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# SGDClassifer() default, loss='hing' -> support vector machine  , 
# Support Vector Machine   Logistic Regressionloss   loss='log' 
model = make_pipeline(StandardScaler(), SGDClassifier(loss='log')) 
model.fit(X_train, y_train) 
```

```python
print("   : {}" .format(model.score(X_train, y_train)))
print("   : {}" .format(model.score(X_test, y_test)))
```

#    SGD 

```python
from sklearn.linear_model import SGDClassifier # SGD 
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# SGDClassifer() default, loss='hing' -> support vector machine  , 
# Support Vector Machine   Logistic Regressionloss   loss='log' 
model = make_pipeline(StandardScaler(), SGDClassifier(loss='log')) 
model.fit(X_train, y_train) 
```

```python
print("   : {}" .format(model.score(X_train, y_train)))
print("   : {}" .format(model.score(X_test, y_test)))
```
