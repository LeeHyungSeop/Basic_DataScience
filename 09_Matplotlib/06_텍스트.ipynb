# 6. 텍스트

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, '-' 마이너스 글자 깨지는 현상 해결

x = [1, 2, 3]
y = [2, 4, 9]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, marker='o')

# y data (2, 4, 9)을 찍어준다 (idx:0,1,2 / txt:2,4,9)
for idx, txt in enumerate(y) : 
    # (x, y+0.3) 위치에 y값을 넣을 것이다
    plt.text(x[idx], y[idx]+0.3, txt, ha='center', color='red') # horizontal arangement : '수평 정렬' 
```
