# 15. 산점도 그래프
두 변수의 상관관계를 보여준다

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, '-' 마이너스 글자 깨지는 현상 해결
```

```python
import pandas as pd

df = pd.read_excel('../08_Pandas_nado/score.xlsx')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np

sizes = np.random.rand(8) * 1000 # 0 ~ 1000 사이의 random값
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes) # 서로 다른 랜덤한 크기의 점들이 찍힌다
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# 1, 2, 3학년 별 점의 크기를 다르게
sizes = df['학년'] * 200 # 1학년 : 200, 2학년 : 400, 3학년 : 600

plt.scatter(df['영어'], df['수학'], s=sizes) 
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# 학년별 점의 크기를 다르게
sizes = df['학년'] * 200 # 1학년 : 200, 2학년 : 400, 3학년 : 600

# 학년별 점의 색깔을 다르게 
# cmap에 해당하는 색깔 : https://matplotlib.org/stable/gallery/color/colormap_reference.html
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(6, 4))

plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.colorbar() 
```

```python
plt.figure(figsize=(6, 6))

plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal') # 1, 2, 3학년에 해당하는 색깔을 알 수 있다. (shrink : colorbar 크기를 작게 하기)
```
