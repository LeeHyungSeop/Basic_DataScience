# 9. 막대 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, '-' 마이너스 글자 깨지는 현상 해결

x = [1, 2, 3]
y = [2, 4, 9]
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184] # 키 데이터

plt.barh(labels, values) # barh (bar horizon -> 수평 막대) 
plt.xlim(175, 195) 
```

## hatch

https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ....
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 200)

for idx, rect in enumerate(bar) :
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center') 
```
