# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 9]
plt.plot(x, y)
# print(plt.plot(x, y))
```

```python
plt.plot(x, y)
plt.show() # [<matplotlib.lines.Line2D at 0x11706af20>] 이러한 불필요한 데이터를 출력 안 할 수 있다.
```

## Title 설정

```python
plt.plot(x, y)
plt.title("Line Graph")
```

```python
plt.plot(x, y)
plt.title("꺾은선 그래프")
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, '-' 마이너스 글자 깨지는 현상 해결
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 list 확인

[f.name for f in fm.fontManager.ttflist] # 사용 가능한 폰트 list 확인 (이름만 출력)
```
