# 16. 여러 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, '-' 마이너스 글자 깨지는 현상 해결

df = pd.read_excel('../08_Pandas_nado/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10)) # 2x3에 해당하는 plot들을 생성
fig.suptitle('여러 그래프 넣기')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 15)) # 2x2에 해당하는 plot들을 생성
fig.suptitle('여러 그래프 넣기')

# (0, 0) == 첫 번째 그래프
axs[0, 0].set_title('첫 번째 그래프') # 제목
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수') # 데이터 설정
axs[0, 0].set(xlabel='이름', ylabel='점수') # x, y축 label 설정
axs[0, 0].set_facecolor('lightyellow') # 전경색
axs[0, 0].grid(linestyle='--', linewidth=0.5)
axs[0, 0].legend(loc='upper center') # 범례

# (0, 1) == 첫 번째 그래프
axs[0, 1].set_title('두 번째 그래프') # 제목
axs[0, 1].plot(df['이름'], df['수학'], label='수학')
axs[0, 1].plot(df['이름'], df['영어'], label='영어')
axs[0, 1].legend()

# (1, 0) == 세 번째 그래프
axs[1, 0].set_title('세 번째 그래프') # 제목
axs[1, 0].barh(df['이름'], df['키'], label='키', alpha=0.8)
axs[1, 0].legend()

# (1, 1) == 네 번째 그래프
axs[1, 1].set_title('네 번째 그래프') # 제목
axs[1, 1].plot(df['이름'], df['사회'], color='green')
```
