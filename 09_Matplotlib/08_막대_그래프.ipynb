```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, '-' 마이너스 글자 깨지는 현상 해결

x = [1, 2, 3]
y = [2, 4, 9]
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184] # 키 데이터

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184] # 키 데이터

plt.bar(labels, values, color='r')
```

```python
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.4) # 각각 막대의 색깔을 다르게, 투명도 0.4
```

```python
plt.bar(labels, values) 
plt.ylim(175, 200)
```

```python
plt.bar(labels, values, width=0.2) # width : bar 두께 조절
```

```python
plt.bar(labels, values, width=0.3) 
plt.xticks(rotation=45) # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=10) # y축의 키 데이터 각도를 10도로 설정
```

```python
ticks = ['1번학생', '2번학생', '3번학생']
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184] 

plt.bar(labels, values)
plt.xticks(labels, ticks) # x축 tick을 이름 데이터 말고 ticks 데이터로 사용하기
```
