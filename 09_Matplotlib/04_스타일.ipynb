# 4. 스타일 

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, '-' 마이너스 글자 깨지는 현상 해결

x = [1, 2, 3]
y = [2, 4, 9]
```

```python
plt.plot(x, y)
```

## 마커(marker)

https://matplotlib.org/stable/api/markers_api.html

```python
plt.plot(x, y, linewidth=10) # line 두께 설정
```

```python
plt.plot(x, y, marker='o') 
```

```python
plt.plot(x, y, marker='o', linestyle='None') 
```

```python
plt.plot(x, y, marker='v', linestyle='None') 
```

```python
plt.plot(x, y, marker='v', markersize=10, linestyle='None') 
```

```python
plt.plot(x, y, marker='x', markersize=10, linestyle='None') 
```

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor='red', linestyle='None') 
```

```python
plt.plot(x, y, marker='X', markersize=20, markeredgecolor='red', linestyle='None', markerfacecolor='yellow') 
```

## 선 스타일

https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, linestyle='-') # 실선은 linestyle 지정을 안해줘도 default이다.
```

## 포맷

```python
plt.plot(x, y, 'go') # color, marker, linestyle
# == plt.plot(x, y, color='g', marker='o', linestyle='None') 
```

## 축약어

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

markerfacecolor : mfc <br>
markersize : ms<br>
markeredgecolor : mec<br>
linestyle : ls<br>

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

## 투명도

alpha : 투명도 (0 ~ 1)

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', alpha = 0.3)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10, 5), dpi=100) # dots per inch (해상도, 확대)
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=10) # dots per inch (해상도, 확대)
plt.plot(x, y)
```
