# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd

# 남자 데이터만 불러오기
'''
1, 2, 3 row에 대해서는 불필요한 정보기 때문에 skip
index는 '행정기관'로 설정
'B : 행정기관' Column과 
'E : 0~4세' ~ 'Y : 100세 이상' Column만 필요

# 201301_201301_연령별인구현황_월간.xlsx
# 202301_202301_연령별인구현황_월간.xlsx
'''
year = 2023
file_name = '202301_202301_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m.head(3)
```

```python
# 숫자에 ','를 없애서 진짜 정수로 만들기
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int) # 1,191,382 --> 1191382 (정수형)
df_m
```

## 여자 데이터 정의

```python

# 여자 데이터만 불러오기
'''
1, 2, 3 row에 대해서는 불필요한 정보기 때문에 skip
index는 '행정기관'로 설정
'B : 행정기관' Column과 
'AB : 0~4세' ~ 'AV : 100세 이상' Column만 필요
'''
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w.head(3)
```

```python
# 0~4세.1 이 나왔던 이유는 앞서 남자 데이터에서도 있었기 때문에 중복되는 데이터에 대해서 .1 을 붙여준 것이다.
df_w.columns = df_m.columns # Column명 통일
df_w.columns
```

```python
# 숫자에 ','를 없애서 진짜 정수로 만들기
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int) # 1,191,382 --> 1191382 (정수형)
df_w
```

## 데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10, 7), dpi=70)
plt.title(str(year)+'년 대한민국 인구 피라미드')
# 남자의 '전국' 인구수 가져오기
plt.barh(df_m.columns, -df_m.iloc[0] // 10000) # 단위 : 만 명
# 여자의 '전구' 인구수 가져오기
plt.barh(df_w.columns, df_w.iloc[0] // 10000) # 단위 : 만 명

plt.savefig(str(year)+'_인구피라미드.png', dpi=100)
plt.show()
```
