# 하이퍼파라미터 선택
* 반복에 의한 확인

```python
from sklearn.datasets import load_iris # scikit-learn의 샘플 데이터 로드를 위해 import
from sklearn.neighbors import KNeighborsClassifier # 분류 알고리
from sklearn.model_selection import train_test_split

iris = load_iris() # sample data load

X = iris.data
y = iris.target

# 특징행렬과 대상벡터를 train용, test용으로 각각 분리시킨다
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=25)
```

```python
# neightbor값을 1~10으로 바꾸면서 훈련용 데이터의 accuracy, 테스트용 데이터의 accuracy를 배열에 저장
train_accuracy = []
test_accuracy = []
neighbors = range(1, 11)

for n in range(1, 11) :
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train, y_train)
    train_accuracy.append(knn.score(X_train, y_train))
    test_accuracy.append(knn.score(X_test, y_test))
```

```python
# neightbor값을 1~10으로 바꾸면서 훈련용 데이터의 accuracy, 테스트용 데이터의 accuracy를 배열에 저장 후 시각화
import matplotlib.pyplot as plt

plt.plot(neighbors, train_accuracy, label='train accuracy')
plt.plot(neighbors, test_accuracy, label='test accuracy')
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.legend();

# test accuracy가 가장 높은 구간이 n=8~10
# 따라서 hyperparameter를 n_neighbors=8 or 9 or 10 으로 하는 것이 모델 성능을 향상시킬 수 있을 것.
```
