# 다중 선형 회귀

* 다중 선형 회귀란 설명 변수인 다수의 특징과 단일 반응 변수 간에 선형 관계가 있다고 가정하고  초평면이라고하는 선형 평면을 이용해 모델링한 것
* 다중 회귀는 설명 변수는 n차원과 반응 변수의 1차원, 모두 n+1개의 차원을 가지며, 초평면은 한 차원 낮은 n차원이 됨
* 다중 선현 회귀를 시각화하려면, 차원 축소나 아이소맵 같은 기술을 사용해야 함

$y = \alpha + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_3 +... + \beta_n x_n$

$y$ : 반응 변수의 예측값<br>
parameter : 알고리즘을 통해 학습하게 되는 모델의 파라미터 <br>
(parameter), $\alpha$ : 절편 항  <br>
(parameter), $\beta$ : 계수 <br>
$x_n$ : 설명 변수


# 다중 선형 회귀의 벡터 정리

  $Y$ : 반응 변수의 예측값을 가진 열 벡터<br>
  $\beta$ : 알고리즘을 통해 학습하게 되는 모델의 파라미터값을 갖는 열 벡터<br>
  $X$ : 설명 변수값을 갖는 m*n차원 행렬. (m : 훈련데이터 개수(행), n : 특징개수(열))<br>

* 선형 회귀 모델의 벡터 표기법 : 
  $Y = X\beta$ <br><br>
  
  
  

* 훈련 데이터로부터 X와 Y의 값을 알고 있으므로 비용함수를 최소화하는 $\beta$를 계산
  
  행렬 연산에서 나누기가 없으므로 역행렬을 사용해야 하고, 역행렬을 사용하려면 square matrix 여야 함.

# 다항 회귀

- 다항 회귀란 반응 변수와 다항식으로 표현된 특징 사이의 선형 관계를 모델링할 수 있는 다중 선형 회귀의 특별한 형태
- 특징을 변환한 후 다중 선형 회귀에서와 같은 방식으로 설명 변수와 반응 변수의 비선형 관계를 모델링함 (차수가 증가하면 곡선 모델(비선형 모델)이 될 때 다항 회귀라고 함)
- 단일 설명 변수의 단일 특징을 사용하고 있지만 모델의 항은 n개


$y$ : 반응 변수의 예측값 <br>
$x^n$ : 설명 변수 (단일 특징이니까 그냥 x. x1 x2 아님)
