## 데이터 제외하기 dronpa

```python
df.dropna() # 전체 데이터 중에서 NaN을 포함하는 데이터 삭제
```

dropna() 옵션
- axis : index, columns (row를 지울지, columns을 지울지 --> default : idnex)
- how : any, all       (any: 결측치가 하나라도 있으면 삭제, all : 전체가 결측치라면 삭제 --> default : any)

```python
df.dropna(axis='index', how='any') # NaN이 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') # NaN이 하나라도 있는 column 삭제
```

```python
df['학교'] = np.nan # 학교 데이터를 모두 NaN으로
df
```

```python
df.dropna(axis='columns', how='all') # 데이터 전체가 NaN인 경우에만 Column 삭제
```
