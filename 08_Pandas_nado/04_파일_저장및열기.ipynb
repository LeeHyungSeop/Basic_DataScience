# 4. 파일 저장 및 열기
DataFrame 객체를 excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd

data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df = pd.DataFrame(data, index=['1번','2번','3번','4번','5번','6번','7번','8번']) 
df.index.name = '지원번호'
df
```

## 저장하기
### csv 파일(.csv)로 저장

```python
df.to_csv('score.csv', encoding='utf-8-sig') # score.csv file에 df 정보 저장 (한글이 꺠지지 않도록 encoding)
```

```python
df.to_csv('score.csv', encoding='utf-8-sig', index=False) # index('지원번호')는 제외하고 저장
```

### 텍스트 파일(.txt)로 저장

```python
df.to_csv('score.txt', sep='\t') # tab으로 구분된 텍스트 파일
```

### 엑셀 파일(.xlsx)로 저장

```python
df.to_excel('score.xlsx')
```

## 열기

### csv 파일(.csv) 열기

```python
df = pd.read_csv('score.csv')
df
```

```python
df = pd.read_csv('score.csv', skiprows=1) # 지정된 갯수 만큼의 row skip하여 가져옴
df 
```

```python
df = pd.read_csv('score.csv', skiprows=[1,3,5]) # 1,3,5 row는 제외(row는 0부터 시작)
df 
```

```python
df = pd.read_csv('score.csv', nrows=3) # 지정된 갯수 만큼의 row만 가져옴
df 
```

```python
# 처음 2 row 무시, 이후에 4 row를 가져옴
df = pd.read_csv('score.csv', skiprows=2, nrows=4)  
df 
```

### 텍스트 파일(.txt) 열기

```python
# 파일 저장할 때, 구분자로 '\t'를 설정했기 떄문에 열 때도 구분자 '\t'로
df = pd.read_csv('score.txt', sep='\t')
df
```

```python
# '지원번호'라는 column을 index로 만들면서 가져온다
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호')
df

''' (위와 똑같은 코드)
df = pd.read_csv('score.txt', sep='\t')
df.set_index('지원번호', inplace=True)
df
'''
```
